# Can we compare GDP's Country between Kuwait and neighboring Gulf Countries

**Data Source**
Arab states of the Persian Gulf   
https://en.wikipedia.org/wiki/Arab_states_of_the_Persian_Gulf

https://data.worldbank.org/indicator/ny.gdp.mktp.cd?locations=KW


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_1345540.csv', skiprows=range(0,3), index_col='Country Name')
df.head()
```

```python
Gulf_Countries = ['Bahrain', 'Kuwait', 'Iraq', 'Oman', 'Qatar', 'Saudi Arabia', 'United Arab Emirates']
Gulf_df = df.loc[Gulf_Countries]
Gulf_df.drop(['Country Code','Indicator Name','Indicator Code','Unnamed: 64'], axis=1, inplace=True)
Gulf_df
```

```python
Gulf_df[[str(x) for x in range(1960, 2020)]] = Gulf_df[[str(x) for x in range(1960, 2020)]]/1000000000
```

```python
Gulf_df
```

```python
Gulf_df = Gulf_df.T
```

```python
plt.style.use('seaborn-colorblind')
Gulf_df.plot(figsize=(14,6), lw=4);
plt.xlabel('Years')
plt.ylabel('GDP in US Dollars $')
plt.title('Kuwait and Neighboring Gulf Countries GDP 1960-2019')
plt.show()
```
